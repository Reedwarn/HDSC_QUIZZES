```python
new_df=df.drop(['date','lights'],axis=1)
new_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
normalised_df=pd.DataFrame(scaler.fit_transform(new_df),columns=new_df.columns)
normalised_df
```

```python
features_df=normalised_df.drop(['Appliances'],axis=1)
features_df
target_df=normalised_df['Appliances']
target_df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features_df,target_df,test_size=0.3,random_state=42)
x_train
```

```python
#To
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
reg.score(x_train,y_train).round(3)
weights=pd.Series(reg.coef_,index=features_df.columns).sort_values()
print (weights)
predicted_linear=reg.predict(x_test)

from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,predicted_linear))
round(rmse,3)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=0.4)
ridge_reg.fit(x_train,y_train)
predicted_ridge=ridge_reg.predict(x_test)

from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,predicted_ridge))
round(rmse,3)
```

```python
from sklearn.linear_model import Lasso
lasso_reg=Ridge(alpha=0.001)
lasso_reg.fit(x_train,y_train)
predicted_lasso=lasso_reg.predict(x_test)
weights=pd.DataFrame(lasso_reg.coef_,index=features_df.columns)
weights
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,predicted_lasso))
round(rmse,3)
```
